### Calculate the scores and convert them to percentages

```python
key = [3, 4, 2, 3, 4, 1]
corrects = pd.DataFrame(0, index=np.arange(len(data)), columns=['score'])

for student in range(0,len(data.index)):
    corrects.loc[student,:] = len([i for i, j in zip(key, data.loc[student, "Q1":"Q6"]) if i == j])


data.insert(7, "score", corrects['score'], True)
data.insert(8, "score_percent", corrects['score']/6, True)

data
```

## Next, check some distributional plots to visualize how the class as a whole scored

```python
font = {'family' : 'Sans',
        'weight' : 'normal',
        'size'   : 18}

matplotlib.rc('font', **font)
```

```python
fig1, axes = matplotlib.pyplot.subplots(2, 2, figsize=(7, 7), sharex=True)

ax = sns.distplot(data['score_percent']*100, kde=False, color="b", ax=axes[0,0])
sns.distplot(data['score_percent']*100, hist=False, rug=True, color="r", ax=axes[0, 1])
sns.distplot(data['score_percent']*100, hist=False, color="g", kde_kws={"shade": True}, ax=axes[1, 0])
sns.distplot(data['score_percent']*100, color="m", ax=axes[1, 1])

ax.set(ylabel='students')
matplotlib.pyplot.tight_layout()
```

```python
fig2, axes = matplotlib.pyplot.subplots(figsize=(14, 6), sharex=True)

ax = sns.violinplot(data=data.loc[:, "Q1":"Q6"], palette="Set3", cut=1)

ax.set(ylabel='answer')
ax.set(yticklabels=['a', 'b', 'c', 'd'])
matplotlib.pyplot.tight_layout()
sns.despine(offset=10, trim=True)
```

```python
fig3, axes = matplotlib.pyplot.subplots(2, 3, figsize=(14, 8), sharex=True)

ax=sns.distplot(data['Q1'], hist=True, bins=8, kde=False, norm_hist=False, color="b", ax=axes[0,0])
sns.distplot(data['Q2'], hist=True, bins=8, kde=False, norm_hist=False, color="b", ax=axes[0,1])

sns.distplot(data['Q3'], hist=True, bins=8, kde=False, norm_hist=False, color="b", ax=axes[0,2])
sns.distplot(data['Q4'], hist=True, bins=8, kde=False, norm_hist=False, color="b", ax=axes[1,0])

sns.distplot(data['Q5'], hist=True, bins=8, kde=False, norm_hist=False, color="b", ax=axes[1,1])
sns.distplot(data['Q6'], hist=True, bins=8, kde=False, norm_hist=False, color="b", ax=axes[1,2])

ax.set(ylabel='students')
ax.set(xticklabels=['a', 'b', 'c', 'd'])
matplotlib.pyplot.tight_layout()
sns.despine(offset=10, trim=True)
```
